```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# In first step we will load the dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# 1. we use Pair Plot which basically shows pairwise relationships
sns.pairplot(df, hue='species')
plt.suptitle('Pairwise Feature Relationships in Iris Dataset', y=1.02)
plt.show()

# 2. Box Plot is used for compare distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal length (cm)')
plt.title('Sepal Length Distribution by Species')
plt.show()

# 3. Violin Plot -> Shows distribution + density
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='species', y='petal width (cm)', inner='point')
plt.title('Petal Width Distribution by Species')
plt.show()

# 4. Heatmap -> Correlation between features
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns='species').corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
